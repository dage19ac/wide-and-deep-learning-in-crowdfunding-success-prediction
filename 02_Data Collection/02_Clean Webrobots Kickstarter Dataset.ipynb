#### Convert Unix Timestamps into Normal Datetime:

```python
print("created_at: {}".format(kickstarter_df.created_at.isna().sum()))
print("deadline: {}".format(kickstarter_df.deadline.isna().sum()))
print("launched_at: {}".format(kickstarter_df.launched_at.isna().sum()))
print("last_update_published_at: {}".format(kickstarter_df.last_update_published_at.isna().sum()))
print("state_changed_at: {}".format(kickstarter_df.state_changed_at.isna().sum()))
```

```python
# Drop attribute last_update_published_at since only NaN values
print("before: {}".format(kickstarter_df.shape))
kickstarter_df.drop(columns=["last_update_published_at"], inplace=True)
print("after: {}".format(kickstarter_df.shape))
```

```python
# Convert attributes from unix timestamp to normal format
from datetime import datetime
kickstarter_df["created_at"] = kickstarter_df.created_at.apply(lambda x: datetime.utcfromtimestamp(x).strftime('%Y-%m-%d'))
kickstarter_df["deadline"] = kickstarter_df.deadline.apply(lambda x: datetime.utcfromtimestamp(x).strftime('%Y-%m-%d'))
kickstarter_df["launched_at"] = kickstarter_df.launched_at.apply(lambda x: datetime.utcfromtimestamp(x).strftime('%Y-%m-%d'))
kickstarter_df["state_changed_at"] = kickstarter_df.state_changed_at.apply(lambda x: datetime.utcfromtimestamp(x).strftime('%Y-%m-%d'))
kickstarter_df.head()
```

```python
# Remove quotation marks from creator name, since they cause a JSON error
import re
def remove_quotation_marks(x):
    name = re.search("(?<=\"name\":\").*?(?=\",)", x).group(0)
    return x.replace(name, "".join(name.split("\"")), 1)

kickstarter_df["creator"] = kickstarter_df.creator.apply(lambda x: remove_quotation_marks(x))
```

```python
# Remove escape characters from middle name, since they cause a JSON error
def remove_escape_chars(x):
    name = re.search("(?<=\"name\":\").*?(?=\",)", x).group(0)
    return x.replace(name, "".join(name.split("\\")), 1)

kickstarter_df["creator"] = kickstarter_df.creator.apply(lambda x: remove_escape_chars(x))
```
