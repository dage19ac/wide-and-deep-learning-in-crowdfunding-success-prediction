#### 1) Retrieve datasets from sqlite3 database and convert them to CSV files

https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/EOYBXM

```python
# Open connection to database
conn = sqlite3.connect("kickstarter.db")
c = conn.cursor()
```

```python
# Show all available tables in database
c.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(c.fetchall())
```

#### a) urls_to_scrape

```python
urls_to_scrape_df = pd.read_sql_query("SELECT * FROM urls_to_scrape", conn)
print(urls_to_scrape_df.shape)
urls_to_scrape_df.head()
```

Result: was probably just used as a to-do list, which urls have to be scraped yet

#### b) all_files

```python
all_files_df = pd.read_sql_query("SELECT * FROM all_files", conn)
all_files_df.set_index("project_id", inplace=True)
print(all_files_df.shape)
all_files_df.head()
```

Result: contains only url and project-id (information will most likely be also available in other tables)
